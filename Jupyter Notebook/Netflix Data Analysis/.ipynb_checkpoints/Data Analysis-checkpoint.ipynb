### Données manquantes

#### Totalités des données manquantes

```python
total_obs_count = data.shape[0]*data.shape[1]
values_nan_count = data.isnull().values.sum()
percentage_values_nan = (values_nan_count / total_obs_count) * 100
print(values_nan_count)
print(percentage_values_nan)
```

Nous avons 4 307 valeurs manquantes dans tout notre DataFrame.

Il y a 4% de données manquantes dans la totalité de nos observations. 

#### Données manquantes pour chaque variable

```python
nan_percentages = (data.isnull().sum() / len(data)) * 100
nan_percentages.plot.bar()
plt.ylabel('Pourcentage de NaN')
plt.xlabel('Variables')
plt.title('Pourcentage de valeurs manquantes par variable')
plt.show()
```

Dans ce graphique à barres, on peut voir que les données manquantes viennent des réalisateurs, de la distribution, du pays et de la date d'ajout de la série ou du film. Il y a une majorité de données manquantes concernant les réalisateurs.

#### Données manquantes par type pour chaque variable

```python
movies_data = data[data["type"] == "Movie"]
tv_shows_data = data[data["type"] == "TV Show"]

nan_percentages = pd.DataFrame({
    'Films': (movies_data.isnull().sum() / len(movies_data)) * 100,
    'Séries': (tv_shows_data.isnull().sum() / len(tv_shows_data)) * 100
})

nan_percentages.plot(kind='bar', figsize=(10, 6))
plt.xlabel('Variables')
plt.ylabel('Pourcentage de valeurs manquantes')
plt.title('Pourcentage de valeurs manquantes des variables par type de contenu')
plt.xticks(rotation=90)
plt.legend()
plt.tight_layout()
plt.show()
```

On aperçoit que les séries ont largement plus de pourcentage de valeurs manquantes que les films.

### Veille sur les différents type de visualisation de données

#### Qu'est ce que la visualisation de données ?

La visualisation des données (ou data visualization ou dataviz en anglais) désigne le fait de représenter visuellement ses data pour pouvoir déceler et comprendre des informations, les données brutes étant difficilement interprétables et exploitables. Ce processus se fait par des outils analytiques spécifiques et se matérialise par des tableaux (type Excel), des graphiques, des cartes visuelles ou même des infographies regroupées dans des dashboards (tableaux de bord).

Voici les visualisations les plus populaires : 
- Diagrammes en secteur
- Tableau (avec textes ou nombres)
- Courbe
- Histogramme (graphique à barres)
- Cartographie
- Infographie
- Frise chronologique

Les visualisations spécifiques : 
- Graphique par aires
- Boîte à moustaches
- Compartimentages
- Nuage de points, de mots ou de bulles
- Vue circulaire
- Arborescence classique ou radiale
- Diagramme de Gantt
- Carte de chaleur
- Matrice
- Cascade
- Entonnoir
- Réseau
- Graphique polaire

#### Librairies de visualisation de données 

Il existe plusieurs bibliothèques python qui sont utilisées pour la visualisation de données dont Matplotlib, Seaborn et Plotly. 

Matplotlib est une bibliothèque largement utilisée pour créer des visualisations statiques, animées et interactives en Python. Il fournit une large gamme de visualisations, y compris des graphiques linéaires, des nuages de points, des graphiques à barres, des histogrammes et des cartes thermiques. Matplotlib fournit également un certain nombre d'options de personnalisation, telles que la modification des couleurs, des styles de ligne et des styles de marqueur. L'un des principaux avantages de Matplotlib est sa capacité à créer des figures de qualité publication, qui peuvent être exportées vers différents formats de fichiers, tels que PNG, PDF et SVG. Cependant, la création de visualisations complexes avec Matplotlib peut prendre du temps et nécessite beaucoup de code.

Seaborn est une bibliothèque construite sur Matplotlib et fournit une interface de niveau supérieur pour créer des visualisations. Seaborn fournit un certain nombre de visualisations supplémentaires, telles que des tracés de violon, des tracés en boîte et des tracés de paires, qui ne sont pas disponibles dans Matplotlib. Seaborn propose également un certain nombre d'options de personnalisation, telles que la modification des couleurs et des styles et l'ajustement automatique de la taille du tracé et du rapport d'aspect. De plus, Seaborn fournit un certain nombre de fonctions pour travailler avec des données statistiques, telles que l'ajustement et la visualisation de modèles de régression linéaire. Le principal avantage de Seaborn est sa capacité à créer des visualisations complexes avec un minimum de code.

Plotly est une bibliothèque permettant de créer des visualisations Web interactives en Python. Plotly fournit une large gamme de visualisations, y compris des tracés linéaires, des nuages ​​de points, des graphiques à barres et des cartes thermiques, et prend en charge un certain nombre de fonctionnalités interactives, telles que le zoom, le panoramique et les étiquettes de survol. Plotly fournit également un certain nombre d'options de personnalisation, telles que la modification des couleurs, des styles de ligne et des styles de marqueur. L'un des principaux avantages de Plotly est sa capacité à créer des visualisations interactives pouvant être visualisées dans un navigateur Web, ce qui facilite le partage de visualisations avec d'autres.

### Visualisation des graphiques

#### Pourcentage des types de contenu

```python
data_country = data["type"].value_counts()

labels = data_country.index
fig, ax = plt.subplots()

ax.pie(data_country, labels=labels, autopct='%1.1f%%', startangle=90)
ax.axis('equal')
plt.title('Répartition des types de contenu sur Netflix')
plt.show()
```

#### Nombre de séries/films par année

```python
number_movies_tv_shows = data.groupby('release_year').size()
plt.plot(number_movies_tv_shows.index, number_movies_tv_shows.values, linestyle='-')
plt.xlabel('Année de sortie')
plt.ylabel('Nombre de séries/films')
plt.title("Évolution du nombre de séries/films au fil du temps")
plt.show()
```

#### Les différentes catégories par année de réalisation

```python
categories = data['listed_in'].str.split(', ', expand=True).stack().unique()

plt.figure(figsize=(10, 10))

for category in categories:
    plt.scatter(data[data['listed_in'].str.contains(category)]['release_year'],
                [category] * len(data[data['listed_in'].str.contains(category)]),
                alpha=0.5)

plt.xlabel('Année de réalisation')
plt.ylabel('Catégories')
plt.title('Catégories vs. Année de réalisation')
plt.yticks(categories)
plt.show()
```

#### Les 10 réalisateurs les plus fréquents

```python
director_counts = data['director'].value_counts().head(10)

plt.figure(figsize=(10, 6))
director_counts.sort_values(ascending=False).plot(kind="bar")
plt.xlabel('Réalisateurs')
plt.ylabel('Nombre d\'apparitions')
plt.title('Top 10 des réalisateurs les plus fréquents')
plt.xticks(rotation=90)
plt.show()
```

#### Les 10 acteurs les plus fréquents

```python
all_actors = []
for actors in data["cast"]
    if isinstance(actors, str):
        all_actors.extend(actors.split(", "))

actor_counts = pd.Series(all_actors).value_counts().head(10)
plt.figure(figsize=(10, 6))
actor_counts.sort_values(ascending=False).plot(kind="bar")
plt.xlabel("Acteur")
plt.ylabel("Nombre d'apparitions")
plt.title("Top 10 des acteurs les plus fréquents")
plt.show()
```

#### Les 5 pays les plus fréquents

```python
all_countries = []
for countries in data["country"]:
    if isinstance(countries, str):
        all_countries.extend(countries.split(", "))

country_counts = pd.Series(all_countries).value_counts().head(5)
plt.figure(figsize=(8, 8))
country_counts.plot.pie(autopct='%1.1f%%', startangle=90)
plt.axis('equal')
plt.title("Répartition des pays les plus fréquents")
plt.show()
```

#### Catégorie les plus fréquentes pour les films

```python
from wordcloud import WordCloud

movies_data = data[data["type"] == "Movie"]
all_categories = []
for categories in movies_data["listed_in"]:
    if isinstance(categories, str):
        all_categories.extend(categories.split(", "))

category_counts = pd.Series(all_categories).value_counts()

wordcloud = WordCloud(width=800, height=800, background_color='white').generate_from_frequencies(category_counts)

plt.figure(figsize=(8, 8))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.title("Nuage de mots des catégories les plus fréquentes pour les films")
plt.show()

```

#### Catégorie les plus fréquentes pour les séries

```python
from wordcloud import WordCloud

tv_shows_data = data[data["type"] == "TV Show"]
all_categories = []
for categories in tv_shows_data["listed_in"]:
    if isinstance(categories, str):
        all_categories.extend(categories.split(", "))

category_counts = pd.Series(all_categories).value_counts()

wordcloud = WordCloud(width=800, height=800, background_color='white').generate_from_frequencies(category_counts)

plt.figure(figsize=(8, 8))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.title("Nuage de mots des catégories les plus fréquentes pour les séries")
plt.show()
```

#### Durée moyenne des films

```python
movies_data = data[data["type"] == "Movie"]

durations = movies_data["duration"].str.extract(r'(\d+)').astype(float)

plt.hist(durations, edgecolor='black')
plt.xlabel('Durée (minutes)')
plt.ylabel('Nombre de films')
plt.title('Distribution des durées des films en minutes')
plt.show()
```

### Durée moyenne des séries

```python
tv_shows_data = data[data["type"] == "TV Show"]

durations = tv_shows_data["duration"].str.extract(r'(\d+)').astype(int)

plt.hist(durations, edgecolor='black')
plt.xlabel('Durée (saisons)')
plt.ylabel('Nombre de séries')
plt.title('Distribution des durées des séries en nombre de saisons')
plt.show()
```

#### Les 10 classifications les plus fréquentes

```python
rating_counts = data['rating'].value_counts().head(10)

rating_counts.sort_values(ascending=False).plot(kind="bar")
plt.xlabel('Classification')
plt.ylabel('Nombre de classifications')
plt.title('Nombre des classifications par âge')
plt.xticks(rotation=90)
plt.show()
```

### Analyse des graphiques

J'ai effectué plusieurs analyses sous forme de différents graphiques (histogramme, diagramme circulaire, diagramme linéaire, nuage de mots, nuage de points) en fonction de l'affichage que je souhaitais obtenir.

Dans un premier temps, j'ai fait un diagramme de la répartition des types de contenu (Série ou Film) sur Netflix, nous pouvons observer qu'il y a une nette différence avec beaucoup plus de films que de séries, ceci peut se traduire par le fait que les films existent depuis plus longtemps (le premier film est sorti en 1845 et la première série en 1946.) soit 100 ans après. Mais également par le fait qu'une série est beaucoup plus longue à produire qu'un film.

J'ai effectué un diagramme linéaire sur l'évolution du nombre de séries ou de films par année. Netflix, étant assez récent, propose des séries et films récents à son audience et moins de contenu très ancien. Netflix a également investi massivement dans la production de contenu. Nous apercevons que la production de films et de séries a énormément augmenté au fil des années.

Au-dessous, sur les différentes catégories par année de réalisation, nous apercevons l'évolution sur les catégories au fil des années. Nous pouvons apercevoir qu'entre 1940 et 1960, les catégories n'étaient pas très variés : film, film classique et documentaires. Et qu'au fil du temps les genres se sont diversifiés. 

Sur les deux graphiques à barre des tops 10 des réalisateurs et acteurs les plus fréquents, nous pouvons voir les 10 réalisateurs et acteurs qui ont produit le plus de film/série.

Pour les pays les plus fréquents, sans grande surprise, nous avons les Etats Unis à la première place, avec presque 60 pourcents des films produits aux Etats-Unis qui sont sur Netflix, Les USA restent le pays emblématique du grand écran.

Nous pouvons voir des nuages de mots des catégories les plus fréquentes pour chaque type de contenu. Les films/series internationaux reviennent le plus, ainsi que les dramas et les comédies. Il n'y a pas de grandes différences entre les films et les séries sur cette répartition. 

Concernant les histogrammes de la durée des films/séries, nous pouvons observer que la plupart des films durent entre 90 et 120 minutes et que les séries ont majoritérement une seule saison. 

Pour finir, un diagramme à barre des nombres de classification par âge, qui montre qu'il y a essentiellement du contenu pour adultes et pour les plus de 14 ans.
